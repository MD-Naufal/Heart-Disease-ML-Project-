# DATA VISUALIZATION FOR HEART DISEASE DATASET

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
file_path = r'C:\Users\moham\Documents\Heart Disease Project\heart.csv'
df = pd.read_csv(file_path)
df.head(10)
```

```python
df.info
```

```python
df.shape
```

```python
# Visualization 1: Age distribution
plt.figure(figsize=(12,5))
sns.histplot(df['Age'], bins=20, kde=True, color="skyblue")
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Visualization 2: Gender distribution
plt.figure(figsize=(12, 5))
sns.countplot(x='Sex', data=df, palette="Paired")
plt.title('Gender Distribution')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()
```

```python
# Visualization 3: Distribution of Cholesterol Levels
plt.figure(figsize=(12, 5))
sns.histplot(df['Cholesterol'], bins=20, kde=True, palette="paired")
plt.title('Cholesterol Level Distribution')
plt.xlabel('Cholesterol')
plt.ylabel('Count')
plt.show()
```

```python
# Visualization 4: Heart Disease vs No Heart Disease
plt.figure(figsize=(12, 5))
sns.countplot(x='HeartDisease', data=df, palette="Paired")
plt.title('Heart Disease Distribution')
plt.xlabel('Heart Disease (1 = Yes, 0 = No)')
plt.ylabel('Count')
plt.show()
```

```python
# Visualization 5: Heart Disease vs No Heart Disease in piechart
heart_disease_count = df['HeartDisease'].value_counts()
labels = ['Heart Disease', 'No Heart Disease']
values = heart_disease_count.values

plt.figure(figsize=(12, 6))
plt.pie(values, labels=labels, autopct='%1.0f%%', startangle=90, colors=['darkblue', 'skyblue'], explode=[0.1, 0], shadow=True)
plt.title('Heart Disease vs No Heart Disease Proportion')
plt.show()
```

```python
# Visualization 6: correlation heatmap 
numerical_df = df.select_dtypes(include=['float64', 'int64'])

plt.figure(figsize=(12, 6))
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Visualization 7: Heart Disease by Age (KDE)
plt.figure(figsize=(12, 5))
sns.kdeplot(df[df['HeartDisease'] == 0]['Age'], label='No Heart Disease', shade=True)
sns.kdeplot(df[df['HeartDisease'] == 1]['Age'], label='Heart Disease', shade=True, color="darkblue")
plt.title('Age Distribution by Heart Disease')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
# Visualization 8: Chest Pain Type distribution
plt.figure(figsize=(12, 5))
sns.countplot(x='ChestPainType', data=df, palette="Paired")
plt.title('Chest Pain Type Distribution')
plt.xlabel('Chest Pain Type')
plt.ylabel('Count')
plt.show()
```

```python
# Visualization 9: Resting Blood Pressure Distribution
plt.figure(figsize=(12, 5))
sns.histplot(df['RestingBP'], bins=20, kde=True, color="skyblue")
plt.title('Resting Blood Pressure Distribution')
plt.xlabel('Resting BP (mm Hg)')
plt.ylabel('Count')
plt.show()
```

```python
# Visualization 10: Exercise Angina Distribution
plt.figure(figsize=(12, 5))
sns.countplot(x='ExerciseAngina', data=df, palette="Paired")
plt.title('Exercise-Induced Angina Distribution')
plt.xlabel('Exercise Angina (Y = Yes, N = No)')
plt.ylabel('Count')
plt.show()
```

```python
# Visualization 11: Maximum Heart Rate Distribution
plt.figure(figsize=(12, 5))
sns.histplot(df['MaxHR'], bins=20, kde=True, color="skyblue")
plt.title('Maximum Heart Rate Distribution')
plt.xlabel('MaxHR (bpm)')
plt.ylabel('Count')
plt.show()
```

```python
# Visualization 12: Oldpeak Distribution by Heart Disease
plt.figure(figsize=(12, 5))
sns.boxplot(x='HeartDisease', y='Oldpeak', data=df, palette="Paired")
plt.title('Oldpeak Distribution by Heart Disease')
plt.xlabel('Heart Disease (1 = Yes, 0 = No)')
plt.ylabel('Oldpeak')
plt.show()
```
