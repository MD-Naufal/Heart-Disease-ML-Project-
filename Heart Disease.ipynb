## Data Preprocessing

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['Sex'] = le.fit_transform(df['Sex'])
df['ChestPainType'] = le.fit_transform(df['ChestPainType'])
df['RestingECG'] = le.fit_transform(df['RestingECG'])
df['ExerciseAngina'] = le.fit_transform(df['ExerciseAngina'])
df['ST_Slope'] = le.fit_transform(df['ST_Slope'])
```

```python
df.head(10)
```

```python
#Exporting label encoding file to folder
df.to_csv(r'C:\Users\moham\Documents\Heart Disease Project\new_file_Heart.csv', index=False)
```

```python
X = df.drop(columns=['HeartDisease'])
y = df['HeartDisease']
print("X Datas",X)
print("y Datas",y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 12)
```

```python
print("DF datas",df.shape)
print("X_train data",X_train.shape)
print("X_test data",X_test.shape)
print("y_train data",y_train.shape)
print("y_test data",y_test.shape)
```

## Model Training

```python
from sklearn.naive_bayes import GaussianNB
NB_data = GaussianNB()
NB_data.fit(X_train, y_train)
```

## Model Evaluation

```python
y_pred = NB_data.predict(X_test)

print("y_pred",y_pred)
print("y_test",y_test)
```

```python
from sklearn.metrics import accuracy_score
print("ACCURACY is", accuracy_score(y_test,y_pred))
```
